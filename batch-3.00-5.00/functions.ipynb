```python
def fibo(n=10,a=0,b=1):
    if a==0 and b==1:
        print(a,b,end=" ")
    for i in range(0,n):
        c=a+b
        a=b
        b=c
        print(c,end=" ")
fibo(15)
```

```python
def fibo(n=10,m=0,a=0,b=1):
    if a==0 and b==1:
        print(a,b,end=" ")
    if m==n:
        return
    c=a+b
    print(c,end=" ")
    fibo(n=n,m=m+1,a=b,b=c)
fibo(15)
```
